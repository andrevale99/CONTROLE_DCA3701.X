# Realimentação de Estado

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import os

A = np.array([
    [1, 0],
    [0, -2]
], dtype=np.float32)

B = np.array([
    [1],
    [1]
], dtype=np.float32)

C = np.array([1,1], dtype=np.float32)

D = 0

K = np.array([-2, 0], dtype=np.float32)

Aa = A + B*K
Ba = B
Ca = C

t = np.linspace(0., 5, 1000)
u = np.zeros_like(t)
x0 = np.array([
    [1],
    [1]
])


# Sistema original (sem realimentação)
system = signal.StateSpace(A, B, C, D)
_, X1, Y1 = signal.lsim(system, U=u, T=t, X0=x0.T)
system = signal.StateSpace(Aa, Ba, Ca, D)
_, X2, Y2 = signal.lsim(system, U=u, T=t, X0=x0.T)

plt.subplot(121)
plt.plot(t, Y1, label="Sem realimentaçao")
plt.legend()


plt.subplot(122)
plt.plot(t, Y2, label="Com realimentaçao")
plt.legend()

plt.show
```
